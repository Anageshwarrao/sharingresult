```r
data <- read.csv("C:\\Users\\LENOVO\\Desktop\\creditcard.csv", header = TRUE,
                  stringsAsFactors = FALSE,
                  strip.white = TRUE ,
                  sep = ',')
#head(data)
```

```r
library("sqldf")
partition1 <- sqldf("select * from data where Class=1")
partition0 <- sqldf("select * from data where Class=0")
nrow(partition1)
```

```r
library("dplyr")
set.seed(123)
partition0 <- sample_n(partition0, 492*3)
nrow(partition0)
```

```r
data2 <- rbind(partition0,partition1)
nrow(data2)
```

```r
set.seed(123)
data2 <- data2[sample(nrow(data2)),]
head(data2)
```

```r
write.csv(data2,file="C:\\Users\\LENOVO\\Desktop\\sampledata.csv")
```
