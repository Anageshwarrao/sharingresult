```r
data1 <- read.csv("C:\\Users\\LENOVO\\Desktop\\sampledata.csv", header = TRUE,
                  stringsAsFactors = FALSE,
                  strip.white = TRUE ,
                  sep = ',')
data1$Class <- as.factor(data1$Class)
```

```r
library("party")
library("partykit")
sample_ctree <- ctree(Class ~ ., data=data1)
print(sample_ctree)
```

```r
plot(as.simpleparty(sample_ctree), type="simple")
```

```r
sample_ctree1 <- ctree(Class ~ V14+V17, data=data1)
print(sample_ctree1)
```

```r
plot(as.simpleparty(sample_ctree1), type="simple")
```

```r
library(tree)
tree1 <- tree(Class ~ ., data = data1)
summary(tree1)
```

```r
plot(tree1)
text(tree1)
```

```r
library(tree)
tree2 <- tree(Class ~ V14+V17, data = data1)
summary(tree2)
```

```r
plot(tree2)
text(tree2)
```
